```python
# Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
# Make a request
'''
We are going to use 'create' function
which generates text by making a request to claude
It contains, model, max tokens and messages which
is a list of messages to send (input to model)

There are 
- user messages
- assistant messages

'''
message = client.messages.create(
    model = model,
    max_tokens = 10, ## Is fun to play with this
    messages = [
        {
            "role":"user",
            "content":"What is a dauschund?"
        }
    ]
)

print(message)
```

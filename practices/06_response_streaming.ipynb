## Streaming 

![alt text](images/streaming.png)

```python
# Create an API client
from anthropic import Anthropic

def add_user_message(messages, text):
    user_message = {"role":"user","content":text}
    messages.append(user_message)


client = Anthropic()
model = "claude-haiku-4-5"
messages = []
add_user_message(messages, "What is a dashchund?")

stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)

for event in stream:
    print(event)
```

```python
# Parse stream
from anthropic import Anthropic

def add_user_message(messages, text):
    user_message = {"role":"user","content":text}
    messages.append(user_message)


client = Anthropic()
model = "claude-haiku-4-5"
messages = []
add_user_message(messages, "What is a dashchund?")

with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="") #Concatenate, show to user
        pass

stream.get_final_message() #For example when want to store in db
```

## Rules
- Request to anthropic should not be made directly from client code
- API requests requires a secret API key, which shouldn't be exposed to the client
- The request to anthropic is always performed by your server

## Your server -> Anthropic api

- You can use SDK or direct HTTP request
- Input text is placed as user

## Model processing
- User input gets tokenized
- Each token is made an embedded
- The meaning of a word depends of the context or words around
- Contextualization, each embedding is adjusted given context to create meaning
- Generation, generates output layer or probabilities for next word
- Before generating, it asks himself to answer.
- Output text plced into assistant message

## Anthropic api -> Your server
- Output text placed into assistant message
- Anthropic api answer to server
- Response data contains the following:
    - Message: Single message that contains the generated text
    - Usage: Number of input + output tokens
    - Stop reason: Why the model stopped the generation

## Your server -> Client
- Server returns the generated text
- We display in screen

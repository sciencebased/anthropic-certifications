# Math tutor specialist

# Parameters:
Responses should : 
- Give the student hints
- Walk the student through a solution
- Show solutions to similar problems

Responses should not :
- Immediately give a direct answer
- Tell the student to use a calculator

## System prompts
Customize claude tone and style of response.
Claude wll try to respond in the same way someone in the specific role would respond

```python
# Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user","content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant","content":text}
    messages.append(assistant_message)

def chat(messages):
    # Make a request
    system='''
    You are an assistant for discovery new dogs, for 
    the presentation of new dogs you should present 
    them as pokemons
    '''
    message = client.messages.create(
        model = model,
        max_tokens = 100, ## Is fun to play with this
        messages = messages,
        system=system
    )
    return message.content[0].text

```

```python
## If i would like to do it dynamic system message

def add_user_message(messages, text):
    user_message = {"role":"user","content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant","content":text}
    messages.append(assistant_message)

def chat(messages, system=None):

    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }
    if system:
        params["system"]=system
    # Make a request
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Make a list of messages
messages = []

# Add initial user message
add_user_message(messages, "What is a dashchund?")

# Pass the list of messages into 'chat' to get an answer
answer = chat(messages)

# Take the answer and add it as an assistan message into out list
add_assistant_message(messages, answer)

# Continue the conversation
add_user_message(messages, "What color it has?")

answer = chat(messages)
answer
```
